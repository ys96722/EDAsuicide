# Cleaning

#### View the column names

```python
data.columns
```

#### Replace the columns names that are hard to read

```python
column_names = ['country', 'year', 'sex', 'age', 'suicides', 'population',
       'suicidesPerK', 'countryYear', 'HDI',
       'gdp', 'gdp_pc', 'generation']
data.columns = column_names
```

#### Dropping the features that will not be covered in this analysis


```python
data = data.drop(columns=['HDI','gdp','gdp_pc'])
```

#### Typecast string-typed variables to category-type

```python
data.year = data.year.astype('category')
data.age = data.age.astype('category')
data.sex = data.sex.astype('category')
data.generation = data.generation.astype('category')
data.country = data.country.astype('category')
data.countryYear = data.countryYear.astype('category')

```

#### Mapping non-integer features to integer values to enable correlation analysis


```python
data['ageInt'] = data['age'].map({
    '5-14 years': 0,
    '15-24 years':1,
    '25-34 years':2,
    '35-54 years':3,
    '55-74 years':4,
    '75+ years': 5
})

data['sexInt'] = data['sex'].map({
    'female': 0,
    'male':1
})

data['genInt'] = data['generation'].map({
    'G.I. Generation': 0,
    'Silent':1,
    'Boomers':2,
    'Generation X':3,
    'Millenials':4,
    'Generation Z': 5
})
```

# Global Trends

### Suicides by Year

```python
suicidesByYear = sns.barplot(data = data, x = 'year', y= 'suicides', estimator=sum, ci =None)
suicidesByYear.set_title('Suicides by Year')
suicidesByYear.set_xticklabels(suicidesByYear.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

Note the sudden decrease at year 2016. I am assuming the data for 2016 is incomplete, so let's remove that data

```python
data.drop(data.index[data['year'] == 2016], inplace = True)
data.year.cat = data.year.cat.remove_unused_categories()
```

### Global Change of Variables over Years

```python
fig, ax = plt.subplots(5, 1, sharex='col', sharey='row')
fig.subplots_adjust(hspace = 0.8)
sns.lineplot(data = data, x = 'year', y= 'population' , ci =None , hue='age', ax=ax[0]).set_title("Global Population Change over Years")
sns.lineplot(data = data, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'generation', hue_order=['G.I. Generation', 'Silent', 'Boomers', 'Generation X', 'Millenials', 'Generation Z'], ax=ax[1]).set_title("Global Rate of Suicides by Generations")
sns.lineplot(data = data, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'sex', hue_order=["male","female"], ax=ax[2]).set_title("Global Rate of Suicides by Gender")
sns.lineplot(data = data, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'age', hue_order=["75+ years", "55-74 years", "35-54 years", "25-34 years", "15-24 years", "5-14 years"],ax=ax[3]).set_title("Global Rate of Suicides by Age Group")
sns.lineplot(data = data, x = 'year', y= 'suicides' , ci =None , hue = 'age', hue_order=["75+ years", "55-74 years", "35-54 years", "25-34 years", "15-24 years", "5-14 years"],ax=ax[4]).set_title("Global Number of Suicides by Age Group")
plt.show()
```

Notable Trends:

1. Looking at the rate of suicide by generation, the older generation stays consistently higher compared to the younger generations. 

2. This is also true for the rate of suicide among age-groups as well. Older age groups consistently have higher rates compared to the younger groups.

3. The NUMBER of suicides has a different trend. The age-group of 35-54 years consistently tops the chart over the years, followed by 55-74 and 25-34 years. This may be due to the fact that the global population in the age of 35-54 years has the highest population throughout the years.

4. Note the gap between the males and females in their rate of suicide.

# Trends in Korea

#### Subsetting the Dataframe

```python
korea = data[data['country'] == 'Republic of Korea']
```

### Change of Variables over Years in Korea

```python
fig, ax = plt.subplots(5, 1, sharex='col', sharey='row')
fig.subplots_adjust(hspace = 0.8)
sns.lineplot(data = korea, x = 'year', y= 'population' , ci =None , hue='age', ax=ax[0]).set_title("Population Change over Years in Korea")
sns.lineplot(data = korea, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'generation', hue_order=['G.I. Generation', 'Silent', 'Boomers', 'Generation X', 'Millenials', 'Generation Z'], ax=ax[1]).set_title("Rate of Suicides by Generations in Korea")
sns.lineplot(data = korea, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'sex', ax=ax[2]).set_title("Rate of Suicides by Gender in Korea")
sns.lineplot(data = korea, x = 'year', y= 'suicidesPerK' , ci =None , hue = 'age', hue_order=["75+ years", "55-74 years", "35-54 years", "25-34 years", "15-24 years", "5-14 years"],ax=ax[3]).set_title("Rate of Suicides by Age Group in Korea")
sns.lineplot(data = korea, x = 'year', y= 'suicides' , ci =None , hue = 'age', hue_order=["75+ years", "55-74 years", "35-54 years", "25-34 years", "15-24 years", "5-14 years"],ax=ax[4]).set_title("Number of Suicides by Age Group in Korea")
plt.show()
```

The trends in the rate of suicide in Korea stays pretty consistent with the global trends. However, there are a few notable differences:

1. The gap between the rate of suicide for 75+ years age-group and the rest is wider compared to the global trend. Inadequate care of elderly population is a known issue in Korea, and the data seems to suggest the similar.

2. The population of uppder age-groups (35 ~ 75+ years) have positive slopes where as the rest of the age-groups are slightly declining over the years. Decrease of numbers in the younger populations is also another known social issue in Korea, and the data seems to suggest the similar.

### Rate of Suicide over Years in Korea

```python
suicideRatesByGenderKorea = sns.lineplot(data = korea, x = 'year',y= 'suicidesPerK', ci =None , hue = 'sex')
suicideRatesByGenderKorea.set_title('Suicide Rates by Gender')
plt.axvline(1997, 0, color='r', linestyle='--', lw=2, label="IMF Crisis")
plt.axvline(2002, 0, color='w', linestyle='--', lw=2, label="2002 World Cup")
plt.axvline(2008, 0, color='b', linestyle='--', lw=2, label="Global Financial Crisis")
plt.axvline(2014, 0, color='y', linestyle='--', lw=2, label="Sinking of Sewol Ship")
plt.legend()
plt.show()
```

### Correlation Heatmap

```python
koreaInt = korea[['year','sexInt','ageInt','genInt','suicides','suicidesPerK', 'population']]
koreaInt['year'] = koreaInt['year'].astype(int)
# fmt= '.1f'
# mask=koreaInt.isnull()

mask = np.zeros_like(koreaInt.corr())
mask[np.triu_indices_from(mask)] = True
koreaHeat = sns.heatmap(data=koreaInt.corr(), annot=True, mask=mask, fmt= '.01f',
                        vmin=-1, vmax=1, linewidths=.5, cmap="YlGnBu", square=True)
```

Notable correlations are:
1. Positive correlation between the population and suicides makes sense. More population a group has, the higher chance for the higher number of suicides. Population's negative correlation with the rate of suicide makes sense for the same reason. 

2. Generation has negative correlation with suicide rates, which means that the younger generations tend to have lower suicide rates.

3. Age has positive correlation with suicide rates, which means that older population tend to have higher suicide rates.
